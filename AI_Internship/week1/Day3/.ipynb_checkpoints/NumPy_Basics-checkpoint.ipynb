# Introduction to NumPy

NumPy (Numerical Python) is the fundamental numerical computing library in Python.

It introduces the ndarray (N-dimensional array), which stores data more efficiently than Python lists and performs mathematical operations significantly faster.

Many popular libraries—including Pandas, Scikit-learn, TensorFlow, and PyTorch—are built on top of NumPy.

## Why NumPy?

Python lists are flexible but become inefficient when working with large numerical datasets.

NumPy solves this by providing:

- Faster numerical computations
- Lower memory usage
- Built-in mathematical operations
- Efficient multidimensional arrays

These features make NumPy the foundation of modern AI and Machine Learning.

## Import NumPy

```python
import numpy as np
```

## NumPy Arrays (ndarray)

The main object in NumPy is called ndarray.

Unlike Python lists, all elements inside a NumPy array share the same data type, allowing operations to be executed much more efficiently.

## Python List vs NumPy Array

```python
python_list = [1, 2, 3, 4]

numpy_array = np.array([1, 2, 3, 4])

print("Python List:", python_list)
print("NumPy Array:", numpy_array)
```

Although they may look similar, NumPy arrays support efficient mathematical operations and consume less memory than Python lists.

## Creating arrays

NumPy provides multiple ways to create arrays depanding on the required structure and values.

Common methods include:
- np.array()
- np.zeros()
- np.ones()
- np.arange()
- np.linspace()
- np.random.rand()

```python
# Creating an array from a python list
a = np.array([1, 2, 3, 4])

# Array filled with zeros
zeros = np.zeros((2, 3))

# Array filled with zeros
ones = np.ones(5)

# Sequence of values - np.arange(start, stop, step)
range_array = np.arange(0, 10, 2)

# Evenly spaced values - np.linspace(start, stop, num)
line_array = np.linspace(0, 1, 5)

# Random values
random_array = np.random.rand(3, 3)

print(a)
print(zeros)
print(ones)
print(range_array)
print(line_array)
print(random_array)
```

## Reshaping Arrays

The reshape() function changes the dimensions of an array without modifying its data.

It is commonly used when preparing datasets for Machine Learning models.

```python
numbers = np.arange(1,17)

matrix = numbers.reshape(4,4)

print(matrix)
```

## Array Attributes

Every NumPy array has important properties:

- **shape**: describes the dimensions of the array
- **dtype**: shows the data type of elements
- **ndim**: shows the number of dimensions
- **size**:  total number of elements

```python
matrix = np.array([[1, 2, 3], 
                  [4, 5, 6]])

print("Shape:", matrix.shape)
print("Data type:", matrix.dtype)
print("Dimensions:", matrix.ndim)
print("Size:", matrix.size)
```

## Indexing and Slicing

### Indexing
Indexing allows accessing a specific element using its row and column indices.

NumPy indexing is similar to Python lists but supports multiple dimensions.

For 1D arrays, indexing works like Python lists.

For 2D arrays:  array[row, column]

Where:
- row represents the row index.
- column represents the column index.

Indexing starts from 0, meaning the first element has index 0.

```python
matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])

print(matrix[0,0])

print(matrix[1,1])

print(matrix[-1,-1])
```

### Slicing

Slicing extracts a subset of an array without modifying the original data.

```python
matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])

print(matrix[:,1])

print(matrix[-1,:])

print(matrix[:1,:2])
```

## Boolean Masking

Boolean masking allows selecting elements based on a condition.

It is commonly used in data filtering and cleaning.

```python
numbers = np.array([1, 2, 3, 4, 5, 6])

mask = numbers > 3

print("Boolean:", mask)
print("Values:", numbers[mask])
```

## Vectorized Operations

Vectorization allows applying operations to entire arrays without writing explicit loops.

It makes code faster, cleaner, and more suitable for large datasets.

```python
a = np.array([1, 2, 3, 4])

print(a * 2)
print(a + a)
print(a.mean())
```

## Broadcasting
