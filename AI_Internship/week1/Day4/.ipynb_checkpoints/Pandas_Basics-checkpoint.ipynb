# Introduction to Pandas

Pandas is a python library designed for data manipulation and analysis.

It provides powerful data structures such as Series and DataFrames, Wich make working with structured data easier.

Pandas is widely used in Data Science workflows before applying Machine Learning models.

# Why Pandas?

Although Python lists can store data, they become difficult to manage when datasets grow larger.

Pandas solves this problem by providing:

- Easy loading of datasets from different file formats.
- Fast manipulation of rows and columns.
- Automatic handling of missing values.
- Powerful filtering and grouping operations.
- Built-in statistical functions.
- Integration with NumPy, Matplotlib, Scikit-learn, TensorFlow, and many other libraries.

Because of these features, Pandas has become the standard library for data analysis in Python.

# Selecting Data from a DataFrame

After loading and inspecting a dataset, the next step is accessing specific parts of the data.

Pandas provides flexible ways to select:

- Individual columns.
- Multiple columns.
- Specific rows.
- Rows that satisfy certain conditions.

Selecting data correctly is essential because almost every data analysis task requires extracting specific information from a dataset.

# Selecting Columns

Columns in a DataFrame represent features or attributes.

Pandas allows selecting columns using their names.

The result depends on whether we select one column or multiple columns.

````python
df["Age"]
````

## Selecting a Single Column

When selecting one column, Pandas returns a Series.

Example:

```python
df["Age"]
```

returns only the Age column with its index.

````python
type(df["Age"])
````

The output is:

`pandas.Series`

because a single column represents a one-dimensional structure.

````python
df[["Age", "Income"]]
````

## Selecting Multiple Columns

To select multiple columns, we pass a list of column names.

Example:

```python
df[["Age", "Income"]]
```

The output is a DataFrame because multiple columns create a two-dimensional table.

# Filtering Rows

Filtering means selecting rows based on a condition.

The condition creates a Boolean mask containing:

- True → keep the row
- False → remove the row

This concept is similar to Boolean masking in NumPy.

````python
df["Age"] > 25
````

The expression above does not return the filtered data directly.

Instead, it creates a Boolean Series that tells Pandas which rows satisfy the condition.

````python
df[df["Age"] > 25]
````

The DataFrame now contains only rows where the Age value is greater than 25.

This technique is one of the most commonly used operations in data analysis.

# Filtering with Multiple Conditions

Sometimes we need more than one condition.

For example:

Select students who are older than 20 and have an income greater than 5000.

````python
df[(df["Age"] > 20) & (df["Income"] > 5000)]
````

## Logical Operators in Pandas

| Operator | Meaning |
|---|---|
| & | AND |
| \| | OR |
| ~ | NOT |

Important:

Each condition must be surrounded by parentheses.

````python
df[(df["Age"] > 25) | (df["Income"] > 7000)]
````

The OR operator (`|`) returns rows where at least one condition is True.

# Selecting Data Using loc and iloc

Pandas provides two powerful indexers:

- `loc` → label-based selection
- `iloc` → position-based selection

They are useful when we need more control over selecting rows and columns.

# loc

`loc` selects data using row and column labels.

The general syntax:

```python
df.loc[row_selection, column_selection]
```

````python
df.loc[0]
````

This returns the row with index label 0.

````python
df.loc[0, "Name"]
````

This selects:

- Row with label 0
- Column "Name"

The result is a single value.

````python
df.loc[df["Age"] > 25, "Income"]
````

This means:

"Select the Income values for rows where Age is greater than 25."

`loc` is very useful when combining filtering with column selection.

# iloc

`iloc` selects data using integer positions.

It works like normal Python indexing.

The syntax:

```python
df.iloc[row_position, column_position]
```

````python
df.iloc[0,1]
````

This selects:

- First row (position 0)
- Second column (position 1)

Remember that indexing starts from zero.

# loc vs iloc

| loc | iloc |
|---|---|
| Uses labels | Uses positions |
| Uses column names | Uses column numbers |
| More readable | Similar to Python lists |
| Good for conditions | Good for fixed positions |

# Best Practices

- Prefer column names instead of positions when possible because they are easier to understand.
- Use `loc` when working with conditions.
- Use `iloc` when you specifically need rows or columns by position.
- Always check column names before selecting data.

# Data Cleaning

Real-world datasets often contain problems such as:

- Missing values.
- Duplicate records.
- Incorrect data types.
- Inconsistent values.

Before performing analysis or building Machine Learning models, these issues must be handled carefully.

Data cleaning improves data quality and ensures more reliable results.

# Why is Data Cleaning Important?

Machine Learning models learn from the data they receive.

If the dataset contains errors, missing information, or incorrect formats, the model may learn wrong patterns and produce inaccurate predictions.

Therefore, cleaning and preparing data is a critical step in every Data Science workflow.

# Handling Missing Values

A missing value means that some information is not available.

For example:

| Name | Age | Income |
|---|---|---|
| Ali | 20 | 5000 |
| Sara | NaN | 6000 |

The missing value is represented by `NaN` (Not a Number).

````python
df.isnull()
````

## isnull()

The `isnull()` function checks every value in the DataFrame.

It returns:

- True → value is missing
- False → value exists

However, for large datasets, viewing the entire output is not practical.

````python
df.isnull().sum()
````

The combination:

```python
df.isnull().sum()
```

is commonly used to count missing values in each column.

It helps identify which columns require cleaning.

# Creating Missing Values for Practice

To understand how missing values are handled, we can create an example.

````python
df.loc[1, "Age"] = None

df
````

````python
df.isnull().sum()
````

# Filling Missing Values

One approach to handle missing values is replacing them with a meaningful value.

For numerical columns, a common choice is the mean (average).

````python
df["Age"].fillna(df["Age"].mean())
````

## fillna()

`fillna()` replaces missing values with a specified value.

Examples:

- Mean → for numerical data
- Median → when data contains outliers
- Mode → for categorical data

Choosing the replacement method depends on the nature of the dataset.

````python
df
````

# Removing Missing Values

Sometimes filling missing values is not the best solution.

If a row contains too much missing information, it may be better to remove it.

````python
df.dropna()
````

## dropna()

The `dropna()` function removes rows containing missing values.

Example:

Before:

| Name | Age |
|---|---|
| Ali | 20 |
| Sara | NaN |

After `dropna()`:

| Name | Age |
|---|---|
| Ali | 20 |


## fillna() vs dropna()

| Method | Usage |
|---|---|
| fillna() | Replace missing values |
| dropna() | Remove missing rows |

The decision depends on the dataset and the importance of the missing information.

# Best Practices for Missing Values

Before handling missing values:

1. Identify how many values are missing.
2. Understand why they are missing.
3. Decide whether to fill or remove them.
4. Document the decision.

Never remove missing data without understanding its impact.

# Grouping Data with groupby()

Often, we need to analyze data based on categories or groups.

For example:

- Average income per department.
- Total sales per product category.
- Maximum score per class.

Pandas provides the `groupby()` function to perform these operations efficiently.

# The Split-Apply-Combine Concept

The idea behind `groupby()` is based on three steps:

1. Split:
   Divide the data into groups based on a column.

2. Apply:
   Perform an operation on each group.

3. Combine:
   Combine the results into a new DataFrame or Series.

This approach is known as:

**Split → Apply → Combine**

Example:

Imagine we have student data:

| Name | Department | Score |
|---|---|---|
| Ali | AI | 90 |
| Sara | CS | 85 |
| Omar | AI | 95 |

We can group students by department and calculate the average score.

````python
data = {
    "Name": ["Ali", "Sara", "Omar", "Lina"],
    "Department": ["AI", "CS", "AI", "CS"],
    "Score": [90, 85, 95, 80]
}

students = pd.DataFrame(data)

students
````

````python
students.groupby("Department")["Score"].mean()
````

## Understanding the Code

```python
students.groupby("Department")["Score"].mean()
```

Step by step:

1. `groupby("Department")`
   
   Creates groups based on department.

2. `["Score"]`

   Selects the column we want to analyze.

3. `.mean()`

   Calculates the average score for each group.

The output shows:

- Average score for AI students.
- Average score for CS students.

This allows us to compare patterns between different groups.

# Common Aggregation Functions

Aggregation means applying a calculation to each group.

Common aggregation functions include:

| Function | Description |
|---|---|
| mean() | Average value |
| sum() | Total value |
| count() | Number of values |
| max() | Maximum value |
| min() | Minimum value |

````python
students.groupby("Department")["Score"].max()
````

````python
students.groupby("Department")["Score"].count()
````

# Multiple Aggregations Using agg()

Sometimes we want to apply different calculations to multiple columns.

The `agg()` function allows us to define several aggregation operations at once.

````python
students.groupby("Department").agg(
    {
        "Score": ["mean", "max", "min"]
    }
)
````

The result contains multiple statistics for each department.

This is useful when performing deeper analysis instead of looking at only one value.

````python
students.groupby("Department").agg(
    {
        "Score": "mean"
    }
)
````

# Practical Example

In real datasets, we might answer questions like:

- What is the average salary for each job category?
- Which product category generates the highest sales?
- What is the maximum value in each group?

GroupBy helps discover patterns and relationships inside data.

# GroupBy Best Practices

- Always understand the meaning of the grouping column.
- Choose aggregation functions that match the analysis goal.
- Check the result after grouping.
- Avoid grouping by columns with too many unique values unless necessary.

## Step 2: Check Missing Values

````python
df.isnull().sum()
````

### Justification

The dataset contains missing values in the **StudyTime** and **Score** columns.

Since both columns contain numerical values, replacing the missing values with the column mean is a reasonable choice. This preserves the number of records while providing an estimated value based on the existing data.

````python
df["StudyTime"] = df["StudyTime"].fillna(df["StudyTime"].mean())
df["Score"] = df["Score"].fillna(df["Score"].mean())
````

````python
df.isnull().sum()
````

# Step 3: Filter the Dataset

In this example, I will filter students who scored more than **80**.

````python
high_score_students = df[df["Score"] > 80]

high_score_students
````

This subset represents high-performing students and could be useful for further performance analysis.

# Step 4: Group and Analyze Data

I will calculate the average score for each course level using `groupby()`.

````python
df.groupby("Level")["Score"].mean()
````

### Interpretation

Grouping the data by course level makes it easier to compare student performance across different learning levels.

This type of analysis helps summarize the data and identify trends within each category.

# Bonus: Data Quality Check

While exploring the dataset, I noticed some data quality issues that are common in real-world datasets.

````python
df["Level"].unique()
````

````python
df["Level"] = df["Level"].str.title()

df["Level"].unique()
````

The values **"Beginner"** and **"beginner"** refer to the same category.

Using `str.title()` standardizes the capitalization and prevents duplicate categories.

````python
df["City"].unique()
````

````python
df["City"] = df["City"].str.title()

df["City"].unique()
````

City names should have consistent formatting.

Standardizing capitalization ensures that cities with different letter cases are treated as the same value.

````python
df[df["Score"] > 100]
````

Exam scores should normally range between 0 and 100.

Values outside this range are considered invalid.

````python
df = df[df["Score"] <= 100]
````

````python
df["Score"].max()
````

````python
df[df["StudyTime"] < 0]
````

Study time cannot be negative.

Negative values most likely result from data entry errors.

````python
df = df[df["StudyTime"] >= 0]
````

````python
df["StudyTime"].min()
````

# Bonus: Data Preprocessing

Before using the dataset for Machine Learning, additional preprocessing steps can improve data quality and consistency.

## Encoding

````python
df["Completed"].unique()
````

````python
df["Completed"] = df["Completed"].map({
    "Yes": 1,
    "No": 0
})
````

````python
df["StudyTime"] = (
    df["StudyTime"] - df["StudyTime"].min()
) / (
    df["StudyTime"].max() - df["StudyTime"].min()
)
````

````python
df["Score"] = (
    df["Score"] - df["Score"].min()
) / (
    df["Score"].max() - df["Score"].min()
)
````

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[["StudyTime", "Score"]] = scaler.fit_transform(
    df[["StudyTime", "Score"]]
)
````
