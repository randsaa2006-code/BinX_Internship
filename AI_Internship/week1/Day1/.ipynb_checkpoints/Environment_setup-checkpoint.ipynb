# Day 1: Data Science Environment Setup

This notebook verifies the setup of a Python environment for AI and Data Science projects.

A reproducible environment ensures that the same libraries and versions can be installed and used across different machines.

## Objectives

- Create a Python virtual environment.
- Install essential data science libraries.
- Verify installed packages.
- Prepare the environment for future AI/ML tasks.

## Python Environment

A virtual environment is an isolated Python environment created for a specific project.

It prevents dependency conflicts between different projects and ensures reproducibility.

## Required Libraries

The main libraries used for data science:

- NumPy: numerical computing
- Pandas: data manipulation and analysis
- Matplotlib: data visualization

```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

## Checking Installed Versions

Checking package versions helps confirm that the required libraries are installed correctly.

Recording versions is important for reproducibility.

```python
print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
print("Matplotlib:", plt.__version__)
```

## Python Information

Checking the Python version used in this environment.

```python
import sys

print(sys.version)
```

## Requirements.txt

The requirements.txt file stores the exact versions of installed packages.

It allows others to recreate the same environment using:

pip install -r requirements.txt
