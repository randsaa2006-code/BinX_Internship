# 1. Core Python Data Types
Python Provides several built-in data types that are used to store different kinds of information.
The most common data types in Data Science include:
- Integer (int)
- Float (float)
- String (str)
- Boolean (bool)
- List
- Tuple
- Dictionary
- Set

understanding these data types is essential because almost every AI or ML project relies on them.

```python
# Integer
age = 19

# Float 
height = 164.5

# String
name = "Rand"

# Boolean
is_student = True

# List: orderd and mutable
skills = ["Python", "AI", "ML"]

# Tuple: ordered and immutable
coordinates = (10, 20)

# Dictionary: Key-value pairs
student = {"Name":"Rand", "Major": "AI"}

# Set: unique values
numbers = {1, 2, 3, 4}

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
print(type(coordinates))
print(type(student))
print(type(numbers))
```

# 2. Control Flow
Control flow detemines how a program makes decisions and repeats operations.
Python mainly use:
- **Conditional statements (if / elif / else)** for decision making.
- **for loops** for iterating collections.
- **while loops** for repeating code until a condition becomes false.

These concepts are frequently used in preprocessing data and implementing algorithms.

```python
# Conditional statements (if/ elif/ else) example:
score = 85

if score >= 90:
    print("Excellent")
elif score >= 70:
    print("Good")
else:
    print("Needs Improvement")
```

```python
# For loop example:
numbers = [1, 2, 3, 4, 5]

for value in numbers:
    if value % 2 == 0:
        print(f"{value} is even")
    else:
        print(f"{value} is odd")
```

```python
# While loop example:
counter = 0

while counter < 5:
    print(counter)
    counter += 1
```

# 3. Functions
A Function is a reusable block of code that performs a specific task.

Advanteges of functions:
- Reduce code duplication.
- Improve readability.
- Make Program easier to maintain.
- Allow the same logic to be reused with different inputs.

A good function usually contains:
- Parameters
- Return values
- Docstring

```python
# Function example:
def normalize(value, min_val, max_val):
    """Scale a value to the 0-1 range."""
    return (value - min_val) / (max_val - min_val)

print(normalize(75, 0, 100)) # 0.75
```

# 4. List Comperhensions
List comperhentions provide a concise and Pythonic way to create lists.
Compared to traditional loops, they:
- Require fewer lines of code.
- Are easier to read.
- Execute efficiently.

They are commonly used when processing datasets.

```python
# Examples of List Comperhensions

# Creating Squares
squares = [x**2 for x in range(10)]
print("Square numbers:", squares)

# Selecting even numbers
evens = [x for x in range(20) if x % 2 == 0]
print("Even numbers:", evens)
```

# 5. Object-Oriented Programming (OOP)
Object-Oriented Programming organizes code into classes and objects.
A class usually contains:
- Attributes -> data
- Methods -> behavior

Although beginners write only simple classes, most Machine Learning libraries such as Scikit-learn and PyTorch are built using OOP.

```python
# OOP example:
class DataPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5
         
p = DataPoint(3, 4)
print(p.distance_from_origin()) # 5.0
```

## Exercise 1:
Create a function that takes a list of numbers and returns:
- mean
- maximum value
- minimum value

The output should be returned as a dictionary.

```python
# Solution:

def statistics(numbers):
    return {
        "Mean" : sum(numbers) / len(numbers),
        "Min" : min(numbers),
        "Max" : max(numbers)}

print(statistics([1, 2, 3, 4, 5]))
```

## Exercise 2:
Rewrite a loop that filters even numbers as a list comperhensions.

```python
# Solution using normal loop

even_numbers = []

for x in range(20):
    if x % 2 == 0:
        even_numbers.append(x)

print("Even numbers:", even_numbers)
```

```python
# Solution using list comperhension

even_numbers = [x for x in range(20) if x % 2 == 0]
print("Even numbers:", even_numbers)
```

## Exercise 3:
Create a class representing a data record.
The class should have:
- At least two attributes.
- At least one method.

```python
# Solution 

class Student:
    def __init__(self, name, major):
        self.name = name
        self.major = major

    def display_info(self):
        print("Student Name:", self.name)
        print("Student Major:", self.major)

st1 = Student("Rand", "AI")
st1.display_info()
```
