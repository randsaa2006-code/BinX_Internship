## Core Python Data Types

Python provides different built-in data types to store and manipulate data.
The most commonly used types in data science are:
int, float, str, bool, list, tuple, dict, and set.

```python
# Integer and Float
age = 19
height = 164.5

# String
name = "Rand"

# Boolean
is_student = True

# List
skills = ["Python", "AI", "Machine Learning"]

# Tuple
coordinates = (10, 20)

# Dictionary
student = {
    "name": "Rand",
    "major": "AI"}

# Set
numbers = {1, 2, 3, 3}

print(type(age))
print(type(student))
print(numbers)
```

## Control Flow

Control flow allows programs to make decisions and repeat tasks.

- if/elif/else are used for conditions.
- for and while loops are used for repetition.

```python
numbers = [1, 2, 3, 4]

for value in numbers:
    if value % 2 == 0:
        print(f"{value} is even")
    else:
        print(f"{value} is odd")
```

## Functions

Functions are reusable blocks of code.
They help organize programs by grouping related operations.

A function can have:
- Parameters: inputs
- Return value: output
- Docstring: explanation of what the function does

```python
def normalize(value, min_val, max_val):
    """Scale a value to the range between 0 and 1."""
    return (value - min_val) / (max_val - min_val)


result = normalize(75, 0, 100)

print(result)
```

## List Comprehensions

List comprehension is a Pythonic way to create lists using a single line of code.
It replaces simple loops and makes code shorter and more readable.

```python
squares = [x**2 for x in range(10)]
even_numbers = [x for x in range(20) if x % 2 == 0]

print(squares)
print(even_numbers)
```

## Object-Oriented Programming (OOP)

OOP organizes code using classes and objects.

A class contains:
- Attributes: data stored in the object
- Methods: functions that describe object behavior

AI libraries such as Scikit-learn and PyTorch use classes heavily.

```python
class DataPoint:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return (self.x**2 + self.y**2)**0.5


point = DataPoint(3, 4)

print(point.distance_from_origin())
```

## Hands-On Lab - Step 1

Create a function that takes a list of numbers and returns:
- mean
- minimum value
- maximum value

The result should be returned as a dictionary.

```python
def statistics(numbers):
    return {
        "mean": sum(numbers) / len(numbers),
        "min": min(numbers),
        "max": max(numbers)
    }


print(statistics([1, 2, 3, 4, 5]))
```

## Hands-On Lab - Step 2

Convert a normal for-loop into a list comprehension.

```python
# Using a loop

even_numbers = []

for x in range(20):
    if x % 2 == 0:
        even_numbers.append(x)

print(even_numbers)


# Using list comprehension

even_numbers = [x for x in range(20) if x % 2 == 0]

print(even_numbers)
```

## Hands-On Lab - Step 3

Create a class representing a data record.
The class should contain:
- At least two attributes
- At least one method

```python
class Student:

    def __init__(self, name, major):
        self.name = name
        self.major = major

    def display_info(self):
        print("Student Name:", self.name)
        print("Student Major:", self.major)
        


student = Student("Rand", "AI")

student.display_info()
```
